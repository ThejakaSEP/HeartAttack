# Plotting the data

```python
numbers = pd.Series(df.columns)
df[numbers].hist(figsize = (20, 20))
plt.show();
```

```python
#Plotting multiple plots with Seaborn
```

```python
fig, axs = plt.subplots(2,3, figsize=(15, 7))

sns.histplot(data=df, x="Age", kde=True, color="skyblue",ax=axs[0,0])
sns.histplot(data=df, x="resting blood pressure", kde=True, color="olive",ax=axs[0,1])
sns.histplot(data=df, x="cholestoral in mg/dl", kde=True, color="gold",ax=axs[0,2])
sns.histplot(data=df, x="maximum heart rate achieved", kde=True, color="teal",ax=axs[1,0])
sns.histplot(data=df, x="Previous peak", kde=True, color="pink",ax=axs[1,1])
```

```python
df.plot(kind = 'density', subplots = True, layout = (5, 5), sharex = False, sharey = False, figsize = (20, 20));
```

```python
#See the correlation of the variables using a Heatmap
corr = df.corr()
f, ax = plt.subplots(figsize = (20, 9))
sns.heatmap(corr, vmax = 1, annot = True);
```
